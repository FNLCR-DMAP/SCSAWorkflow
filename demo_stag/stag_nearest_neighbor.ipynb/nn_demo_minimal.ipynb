## 2.Â (Option) Load an `AnnData` object

```python
# Comment this cell if `adata` is already in memory
with open("transform_output.pickle", "rb") as f:
    adata = pickle.load(f)
print(f"AnnData loaded â†’ {adata.n_obs} cells, {adata.n_vars} variables")
```

## 3â€‘4.Â Compute NN + assign pin colors

```python
annotation = "renamed_clusters"
distance_key = "spatial_distance"

# Compute distances
calculate_nearest_neighbor(adata, annotation=annotation,
                           spatial_associated_table="spatial",
                           imageid=None, label=distance_key)

# Simple pinâ€‘color map
import matplotlib.cm
phenotypes = sorted(adata.obs[annotation].unique())
cmap = matplotlib.cm.get_cmap("tab20", len(phenotypes))
adata.uns['pin_colors'] = {p: matplotlib.colors.to_hex(cmap(i))
                           for i, p in enumerate(phenotypes)}

print("Nearest neighbors stored and pin colors assigned.")
```

## 5.Â Quick visualization

```python
result = visualize_nearest_neighbor(
    adata=adata,
    annotation="renamed_clusters",
    spatial_distance="spatial_distance",
    distance_from="Follicular helper T cells (Tfh)",
    distance_to=None,          # All other phenotypes
    plot_type="violin",
    log=True,
    palette=adata.uns['pin_colors'],
)
plt.show()
```
